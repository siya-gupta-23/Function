```python
# 5. What are iterators in Python and how do they differ from iterables?

# 1. Iterables

# An iterable is any Python object that can return an iterator, i.e., an object that can be looped over (e.g., using a for loop). Common examples include lists, tuples, strings, dictionaries, and sets.

# An iterable is an object that implements the __iter__() method, which returns an iterator.
# You can use a for loop to traverse through its elements.

#Example of an Iterable:

# A list is an iterable
my_list = [1, 2, 3, 4]

# You can iterate over it using a for loop
for item in my_list:
    print(item)
# 2. Iterators
Definition: An iterator is an object that represents a stream of data; it produces the items of an iterable one at a time. An iterator object implements two methods: __iter__() and __next__().
Characteristics:
The __iter__() method returns the iterator object itself.
The __next__() method returns the next item in the sequence. When there are no more items to return, it raises a StopIteration exception.
Iterators are "consumed" as they are used. Once you have iterated through the elements, the iterator cannot be reset or reused unless explicitly done so.
Example of an Iterator:
python
Copy code
# Create an iterator from an iterable (e.g., a list)
my_list = [1, 2, 3, 4]
my_iterator = iter(my_list)

# Use the next() function to get items one at a time
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
print(next(my_iterator))  # Output: 4
# Calling next() again would raise StopIteration since there are no more items
Key Differences:
Creation:

Iterable: Any object with an __iter__() method or one that defines both __getitem__() (for index-based access) and __len__() methods is considered iterable.
Iterator: An iterator is created from an iterable using the iter() function or by implementing the __iter__() and __next__() methods in a custom object.
Usage:

Iterable: You can pass an iterable to a for loop to iterate over its elements. An iterable can also be passed to the iter() function to get an iterator.
Iterator: You can manually get each item from an iterator using the next() function until a StopIteration exception is raised.
```

```python
# 6. Explain the concept of generators in Python and how they are defined.

# Generators in Python are special type of functions that allow you to create an iterable sequence of values. 
#A generator function returns a generator object, which can be used to generate the values one-by-one as you iterate over it. 
#Generators are a powerful tool for working with large or complex data sets.
#they allow you to generate the values on-the-fly, rather than having to create and store the entire sequence in memory.
#In Python, you can create a generator by using the yield statement in a function. 
#The yield statement returns a value from the generator and suspends the execution of the function until the next value is requested.

#Example-->

def my_generator():
    for i in range(5):
        yield i
gen = my_generator()
print(next(gen))
print(next(gen))
```

```python
#10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python? 


# 1. map():
#Purpose: Applies a given function to each item of an iterable (e.g., a list) and returns a map object (an iterator) with the results.
#Usage: When you want to transform each element in a list or other iterable.
#Example:

# Function to square a number
def square(x):
    return x * 2

# List of numbers
numbers = [1, 2, 3, 4]

# Apply square function to each element in the list
result = list(map(square, numbers))

print(result)  # Output: [2, 4, 6, 8]

#2. filter():

#Purpose: Filters the elements of an iterable, returning only those that satisfy a given condition (i.e., for which the function returns True).
#Usage: When you want to select only certain elements from a list based on a condition.

#Example:
# Function to check if a number is even
def is_even(x):
    return x % 2 == 0

# List of numbers
numbers = [1, 2, 3, 4, 5, 6]

# Filter out only even numbers
result = list(filter(is_even, numbers))

print(result)  # Output: [2, 4, 6]

#3. reduce():

#Purpose: Applies a function to the items of an iterable cumulatively, from left to right, so as to reduce the iterable to a single value.
#Usage: When you need to combine all elements of a list into a single result (e.g., summing all elements).

#Example:
from functools import reduce

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Calculate the sum of the numbers using the reduce function
sum = reduce(lambda x, y: x + y, numbers)

# Print the sum
print(sum)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
 
def is_prime(num):
   if num <= 1:
       return False
   for i in range(2, num):
       if num % i == 0:
           return False
   return True

 # Check and print prime numbers from 1 to 200
 for n in range(1, 201):
   if is_prime(n):
       print(n, "is a prime number")
   else:
       print(n, "is not a prime number")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

class FibonacciIterator:
  

  def __init__(self, count):
      self.count = count
      self.a = 0
      self.b = 1
      self.n = 0

  def __iter__(self):
      return self

  def __next__(self):
      
      if self.n < self.count:
          self.n += 1
          if self.n == 1:
              return self.a
          elif self.n == 2:
              return self.b
          else:
              c = self.a + self.b
              self.a = self.b
              self.b = c
              return c
      else:
          raise StopIteration  

terms = 10  
fib_iterator = FibonacciIterator(terms)
for number in fib_iterator:
  print(number)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def expo(n):
  for i in range(n+1):
       yield 2**i
    
n= int(input("Enter the number for you want to find the power of 2"))
for i in expo(n):
  print(i)
```
